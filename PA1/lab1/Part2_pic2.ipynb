```r
library(ggplot2)
library(reshape2)

file1 <- "InputData/WHO_NREVSS_Public_Health_Labs.csv"
# get the headers
headers <- read.csv(file1, skip = 1,  header = F, nrows = 1, as.is = T)
# get the rest of columns without headers
df <- read.csv(file1, skip = 2, header = F, sep = ",")

# combine hearders with data
colnames(df)= headers
df$WEEK <- formatC(df$WEEK, width = 2, flag="0")
df$WEEK <- paste(df$YEAR, df$WEEK)
df <-as.data.frame(apply(df,2,function(x)gsub('\\s+', '',x)))
drops <- c("REGION_TYPE", "REGION", "YEAR", "TOTAL_SPECIMENS" )
df <- df[, !(names(df)) %in% drops]
                         
# make row name
rownames(df) <- factor(df$WEEK)
# remove the data after Jan 27th, week4
df <- df[1:(nrow(df)-2),]
drop <- c("WEEK")
table <- df[, !(names(df)) %in% drop]
table <- as.matrix(t(table[, c("BYam", "BVic", "B", "H3N2v", "A(H3)", "A(2009 _H1N1)", "A(Subtyping not Performed)")]))
```

```r
opar <- par(no.readonly=TRUE)
# add extra space to right margin of plot within frame
# c(bottom, left, top, right)
par(mai=c(1,.5, 1, .2))
par(mar=c(3, 2, 4, 2) + 0.1)

# side: side of the graph to draw the axis (1=bottom, 2=left, 3=top, 4=right).

barplot(table, xlab="Week", ylab = "Number of Positive Spocimons",
        col=c("chartreuse1","darkolivegreen1", "chartreuse4","darkorchid4","red","darkorange2", "yellow1"),border="black", font.lab=2)
title(main = "Influenza Positive Tests Reports to CDC by U.S. Public Health\n Laboratories, National Summary, 2017-2018 Season")
legend("topleft", inset=0.006, rownames(table), fill=c("chartreuse1","darkolivegreen1", "chartreuse4","darkorchid4","red","darkorange2", "yellow1"))
```
