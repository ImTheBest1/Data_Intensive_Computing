```r
library("dplyr")
library("tibble")
```

```r
file1 <- "InputData/Genetic04.csv"
df <- read.csv(file1,sep = ",")
colnames(df)[1] <- "type"
colnames(df)[2] <- "group"
colnames(df)[3] <- "number"
colnames(df)[4] <- "total"
p1 <- filter(df, df$type == "H3")
p2 <- filter(df, df$type == "H1pdm09")
p3<- filter(df, df$type == "B/Victoria")
p4<- filter(df, df$type == "B/Yamagata")
```

```r

#main pie chart
data <- c(18068, 1896, 348, 228, 2292, 921)
label <- c("Influenza A (H3N2)", "Influenza A (H1N1) pdm09", "Influenza A (substype unknown)", "Influenza B Victoria",
           "Influenza B Yamagata", "Influenza B (Lineage not determined)")
colors <- c("red","chocolate2", "gold", "lawngreen", "greenyellow", "olivedrab4")
df = data.frame(label, data)

pie(data,labels = data, col=colors,radius = 0.8,
   main = "Influenza Positive Specimens Reported by\n  U.S. Public Health Laboratories,\n  Cumulative, 2017 - 2018 Season")
legend("bottomleft", label, cex = 0.8, fill = colors, bty = "n")
```

```r
#make 2*2 grid to store 4 pies
par(mfrow=c(2,2))
#influenza A (H3N2)

pie.slices <- p1$number
pie.slices <- pie.slices[ pie.slices != 0 ]
lbls <- p1$group
lbls <- lbls[p1$number != 0]
pct <- round(pie.slices/sum(pie.slices)*100) 
lbls <- paste(lbls,"\n") 
lbls <- paste(lbls,pie.slices) 
lbls <- paste(lbls, pct) # add percents to labels 
lbls <- paste(lbls,"%",sep="") # ad % to labels
pie(pie.slices,labels = lbls, col=c("plum1","red", "seashell"),main="Influenza A (H3N2)")

# Influenza A(H1N1)pdm09
slices <- p2$number
slices <- slices[ slices != 0 ]
lbls <- p2$group
lbls <- lbls[p2$number != 0]
pct <- round(slices/sum(slices)*100)
lbls <- paste(lbls,"\n")
lbls <- paste(lbls,slices) 
lbls <- paste(lbls, pct) # add percents to labels 
lbls <- paste(lbls,"%",sep="") # ad % to labels
pie(slices,labels = lbls, col=c("orange"),main="Influenza A(H1N1)pdm09")

#Influenza B Victoria
slices <- p3$number
slices <- slices[ slices != 0 ]
lbls <- p3$group
lbls <- lbls[p3$number != 0]
pct <- round(slices/sum(slices)*100)
lbls <- paste(lbls,"\n")
lbls <- paste(lbls,slices) 
lbls <- paste(lbls, pct) # add percents to labels 
lbls <- paste(lbls,"%",sep="") # ad % to labels
pie(slices,labels = lbls, col=c("olivedrab1", "lightcyan"),main="Influenza B Victoria")

# Influenza B Yamagata
slices <- p4$number
slices <- slices[ slices != 0 ]
lbls <- p4$group
lbls <- lbls[p4$number != 0]
pct <- round(slices/sum(slices)*100)
lbls <- paste(lbls,"\n")
lbls <- paste(lbls,slices) 
lbls <- paste(lbls, pct) # add percents to labels 
lbls <- paste(lbls,"%",sep="") # ad % to labels
pie(slices,labels = lbls, col=c("olivedrab1", "lightcyan"),main="Influenza B Yamagata")
```
