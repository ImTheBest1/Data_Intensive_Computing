```r
ss.data = data.frame(
    pop=c("E1", "E2", "E3", "E4", "E5", "E6", "E7", "C1", "C2", "C3", "C4"),
    obs=c(0.0027, 0.0018, 0.0464, 0.0095, 0.0034, 0.0117, 0.017, 0.1178,
          0.0449, 0.039, 0.0903),
    proc=c(0.0319, 0.0196, 0.0511, 0.0143, 0.0048, 0.0078, 0.0396, 0.1662,
           0.074, 0.1681, 0.1358), stringsAsFactors=FALSE)
ss.data
```

```r
# Add new column 'species' by removing the trailing digits from 'pop'.
ss.data$species = gsub("\\d", "", ss.data$pop)
ss.data
```

```r
# Convert data to long-form with 'melt' from the reshape2 package.
mdat = melt(ss.data, id.vars=c("pop", "species"),
            measure.vars=c("obs", "proc"))
mdat
```

```r
plot_1 = ggplot(mdat, aes(x=pop, y=value, fill=variable)) +
         geom_bar(position="stack", stat="identity") +
         scale_fill_manual(values=c("grey50", "grey80"))
plot_1
ggsave("plot_1.png", plot=plot_1, width=6.5, height=4)
```

```r
rm(list = ls())
library(ggplot2)
library(grid)
library(dplyr)
library(reshape2)
database <- read.csv("national/WHO_NREVSS_Clinical_Labs.csv", header = TRUE, skip = 1)
database$WEEK <- formatC(database$WEEK, width = 2, flag="0")
# change col name TOtal_A to A
colnames(database)[6] <- "A"
colnames(database)[7] <- "B"

#extract and melt
database <- transform(database, WEEK=interaction(YEAR,WEEK,sep=''))
new_dataset <- database[,c(4,6,7)]
new_dataset
new_dataset <- melt(new_dataset, id.vars='WEEK')
new_dataset
new_dataset1 <- database[,c(4,8,9,10)]
new_dataset1
new_dataset1 <- rbind(new_dataset1,new_dataset1)
merged_data <- cbind(new_dataset,new_dataset1)
merged_data

merged_data



cbbPalette <- c("#F0E442","#009E73")  #set color
##plot1
g1 <- ggplot(merged_data, aes(as.character(WEEK), value, fill = variable)) + 
  geom_bar(stat='identity', colour = "black") + 
  scale_fill_manual(values = cbbPalette) + 
  xlab("Week") +
  ylab("Number of Positive Specimens") + 
  ggtitle("Influenza Positive Tests Reported to CDC by U.S. Clinical Liboratories, National Summary, 2017-2018 Season")

g1 <- g1 + geom_line(aes(y=PERCENT_POSITIVE*700, group = 1, colour = "Percent Positive")) + 
  geom_line(aes(y=PERCENT_A*700, group = 1, colour = "% Positive Flu A"), linetype = "dotted", size = 1.2) + 
  geom_line(aes(y=PERCENT_B*700, group = 1, colour = "% Positive Flu B"), linetype = "dotted", size = 1.2) +
  scale_colour_manual(values=c("gold3","springgreen4","black"))

g1 <- g1 + scale_y_continuous(sec.axis = sec_axis(~.*0.0014, name = "Percent Positive")) +
  theme_classic()
g1
```

```r
library(ggplot2)
library(dplyr)

us <- map_data("state")

arr <- USArrests %>% 
  add_rownames("region") %>% 
  mutate(region=tolower(region))

gg <- ggplot()
gg <- gg + geom_map(data=us, map=us,
                    aes(x=long, y=lat, map_id=region),
                    fill="#ffffff", color="#ffffff", size=0.15)
gg <- gg + geom_map(data=arr, map=us,
                    aes(fill=Murder, map_id=region),
                    color="#ffffff", size=0.15)
gg
gg <- gg + scale_fill_continuous(low='thistle2', high='darkred', 
                                 guide='colorbar')
gg <- gg + labs(x=NULL, y=NULL)
gg <- gg + coord_map("albers", lat0 = 39, lat1 = 45) 
gg <- gg + theme(panel.border = element_blank())
gg <- gg + theme(panel.background = element_blank())
gg <- gg + theme(axis.ticks = element_blank())
gg <- gg + theme(axis.text = element_blank())
gg
```
