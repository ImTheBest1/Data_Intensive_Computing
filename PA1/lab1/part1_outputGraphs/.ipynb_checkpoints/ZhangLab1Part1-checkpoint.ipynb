### problem1:  Define two synthetic vectors of data representing sales over 12 months for 2 items. Compare the two using lines graphs. (Discussion; Run the code multiple time and see randomness of the second set of data..)

```r
sales1 <- c(12,14,16,29,30,45,19,20,16,19,34,20)
sales2 <- rpois(12,34)  # random numbers, poisson distribution, mean at 34,12 numbers par
print (sales2)

# bg == background
par(bg = "cornsilk")
# type is the option apply to the graph, 
# types reference website: https://www.statmethods.net/graphs/images/lines0.png
# Plotting Symbols, pch == print character, can use with type
# plot(sales1, col="blue", type = "o", pch = 2, ylim=c(0,100), xlab="Month", ylab="Sales" )
plot(sales1, col="blue", type="o", ylim=c(0,100), xlab="Month", ylab="Sales" )
title(main="Sales by Month")

# line type: lty, range from 1-6
lines(sales2, type="o", pch=22, lty=2, col="red")
# grid: https://stat.ethz.ch/R-manual/R-devel/library/graphics/html/grid.html
# grid(nx = NULL, ny = NULL)
grid(nx=NA, ny=NULL)
# grid(6, 6)
legend("topright", inset=.05, c("Sales1","Sales2"), fill=c("blue","red"), horiz=TRUE)
```

### Problem 6: Accessing external APIs

```r
library("ggmap")

install.packages("sp")
library(sp)

install.packages("maptools")
library(maptools)

install.packages("maps")
library(maps)
```

```r
visited <- c("SFO", "Chennai", "London", "Melbourne", "Johannesbury, SA")

ll.visited <- geocode(visited)
# a <- geocode(visited)

visit.x <- ll.visited$lon
visit.y <- ll.visited$lat

map("world", fill=TRUE, col="white", bg="lightblue", ylim=c(-60, 90), mar=c(0,0,0,0))
points(visit.x,visit.y, col="red", pch=36)
```

```r
library("ggmap")
library("maptools")
library(maps)
cities <- c("SFO", "New York", "Buffalo", "Dallas, TX")
ll.cities <- geocode(cities)
visited.x <- ll.cities$lon
visited.y <- ll.cities$lat
map("state", fill=TRUE, col=rainbow(50), bg="lightblue", mar=c(0,0,0,0))
points(visited.x,visited.y, col="yellow", pch=36)
```

### Problem 8: Working with ggplot2 package (), loading a package, installing package. Object- oriented and incremental additions (extensibility) are special features of this package. We can layer the commands to a base plot

```r
# reference: https://www.rdocumentation.org/packages/ggplot2/versions/2.2.1/topics/aes
# loading a package
library(ggplot2)
```

```r
# aes: List of name value pairs giving aesthetics to map to variables
# The point geom is used to create scatterplots. geom_point in ggplot2
ggplot(mpg, aes(displ, hwy)) + geom_point(mapping =aes(displ, hwy) ) # all are the same
ggplot(mpg, aes(displ, hwy)) + geom_point(mapping =NULL )
ggplot(mpg, aes(displ, hwy)) + geom_point()
```
